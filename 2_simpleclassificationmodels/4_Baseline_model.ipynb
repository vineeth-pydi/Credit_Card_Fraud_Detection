## Training, Testing and Evaluation of the model 

```python
# Training and testing the baseline model on the training data using StratifiedK-fold

## Creating the X (estimator) and y (target variable)
X = df_train.drop(columns = ['isFraud'])
y = df_train['isFraud']

## Stratified K-fold
"""We opted for stratified K-fold because it addresses the imbalance present in the dataset by ensuring that 
the proportions of both classes remain consistent in both the train and test sets. Additionally, given the limited 
data for one class and the numerous features, different splits could result in varying accuracy or chosen metrics. 
The K-fold method enhances model stability by considering multiple splits and computing the evaluation metric for each split."""

numFolds = 3
skf = StratifiedKFold(n_splits=numFolds)

metrics={}
metrics['baseline']=[]
train_ids = []
dev_ids = []
for train, dev in skf.split(X, y):
  y_train = y[train]              # takes the indices of the split and assigns it to train 
  y_dev = y[dev]                  # takes the indices of the split and assigns it to dev

  y_baseline_train = np.zeros(len(y_train))      # Assigning the majority class to all the y train sample
  y_baseline_dev = np.zeros(len(y_dev))          # Assigning the majority class to all the y dev samples

  train_baseline_acc,train_baseline_f1,train_baseline_prec,train_baseline_rec,train_baseline_rocauc = \
  computeModelMetrics(y_baseline_train,y_baseline_train,y_train)     # no data to train the model so pass in the baseline values 
                                                                     # for y pred, y pred proba and compare with actual y train
  dev_baseline_acc,dev_baseline_f1,dev_baseline_prec,dev_baseline_rec,dev_baseline_rocauc = \
  computeModelMetrics(y_baseline_dev,y_baseline_dev,y_dev)

  baseline_metrics_df = createMetricsDF(train_baseline_acc,train_baseline_f1,
                                        train_baseline_prec,train_baseline_rec,train_baseline_rocauc,
                                        dev_baseline_acc,dev_baseline_f1,
                                        dev_baseline_prec,dev_baseline_rec,dev_baseline_rocauc)
  train_ids.append(train)
  dev_ids.append(dev)
  
  metrics['baseline'].append(baseline_metrics_df)

train_dev_indices = {}
train_dev_indices['train'] = train_ids
train_dev_indices['dev'] = dev_ids
# train_dev_indices_df = pd.DataFrame.from_dict(train_dev_indices)
# train_dev_indices_df.to_csv('train_dev_indices_' + str(numFolds) + '.csv',sep=',',index=False)
```

```python
# Saving the split indices (the same idices can be used for all models)
# pickle is a file type in python
import pickle

fileName = 'train_dev_indices_' + str(numFolds) +'.pickle'
with open(fileName, 'wb') as handle:
    pickle.dump(train_dev_indices, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

## Summary

In this notebook, we constructed the baseline model, which serves as the benchmark for evaluating the performance of subsequent models. The baseline model simply predicts the majority class (non-fraudulent transactions) for all instances. Consequently, it achieves an accuracy score of approximately 96%, reflecting the high prevalence of non-fraudulent transactions in the dataset.

For any other model to be considered successful, it must surpass this baseline accuracy and also exhibit a high recall score and a low precision score. Given the highly imbalanced nature of the data, the evaluation metrics chosen include ROC-AUC score, recall score, and accuracy, with a focus on ensuring that the model outperforms the baseline.

Recall holds particular importance in this project as it indicates the model's ability to correctly identify all fraudulent transactions within the dataset. Mathematically, recall is defined as the number of true positives divided by the sum of true positives and false negatives. Therefore, as the number of false negatives increases or the number of true positives decreases, the model's recall score decreases.

Since it's critical not to miss any fraudulent transactions, achieving a high recall score, along with high ROC-AUC and accuracy scores exceeding the baseline, is imperative for ensuring the model's reliability, given the complexity of the dataset.
